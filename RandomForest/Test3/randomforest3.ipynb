# Xử lý dữ liệu

Tất cả dữ liệu ở đây đều là dạng cột nên ta chỉ cần xử lý các giá trị NaN

```python
df.fillna(df.median(), inplace=True)
```

# Xây dựng mô hình Random Forest

Chia dữ liệu thành 2 tập train, test với tỉ lệ 8:2, ở đây giá trị cần dự đoán là mức giá trong cột price_range

```python
# Chia dữ liệu train, test

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.2, random_state=42)

x_train = train.drop(columns=['price_range'])
y_train = train['price_range']

x_test = test.drop(columns=['price_range'])
y_test = test['price_range']
```

Tạo và huấn luyện mô hình

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_jobs = -1)
RF.fit(x_train, y_train)
```

```python
acc = RF.score(x_test, y_test)
print(acc*100)
```
